```python
from plotly.offline import init_notebook_mode, download_plotlyjs
import plotly.graph_objs as go
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
```

```python
import pandas as pd
```

```python
ucdataset=pd.read_csv('dataset.csv')
ucdataset.drop(ucdataset.index[7])
ucdataset.head()
```

```python
ucdataset.drop_duplicates()
ucdataset.head()
```

```python
import numpy as np
ucdataset['Rating'] = np.where((ucdataset.Rating >=4.07),1,ucdataset.Rating)
ucdataset['Rating'].mask(ucdataset.Rating !=1, 0, inplace=True)
```

```python
cdataset=ucdataset[['Fingerprint Sensor','Rating']]
cdataset
```

```python
cdataset['Fingerprint Sensor']=cdataset['Fingerprint Sensor'].map({'yes':7.1, 'No':7.2, 'NaN':7.3})

cdataset
```

```python
x=cdataset.reindex(['Fingerprint Sensor'])
x=x.transpose()
y=cdataset.reindex(['Rating'])
y=y.transpose()
print(x.shape,y.shape)
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(x, y, test_size = 0.2)
print('Training Samples: {}, Test Samples: {}'.format(Xtrain.shape[0], Xtest.shape[0]))
```

```python
sc_X = MinMaxScaler()

sc_X.fit(Xtrain.loc[:, :])
Xtrain.loc[:, :] = sc_X.transform(Xtrain.loc[:, :])

sc_X.fit(Xtest.loc[:, :])
Xtest.loc[:, :] = sc_X.transform(Xtest.loc[:, :])

clf = SVC(probability = True, C = 0.5, gamma = 15)
clf.fit(Xtrain, Ytrain)
```
